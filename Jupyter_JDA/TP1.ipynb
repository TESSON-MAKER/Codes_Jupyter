# Consignes pour le TP : Visualisation des Notes avec Python

## Objectif :
Créer un Jupyter Notebook qui calcule des statistiques sur une liste de notes et affiche des graphiques pour visualiser les résultats.

## Étape 1 : Préparation de l'environnement

**Ouvrir Jupyter Notebook :**
- Lancez Jupyter Notebook depuis votre terminal ou Anaconda Navigator.
- Créez un nouveau notebook en cliquant sur "New" > "Python 3".

**Importer les bibliothèques nécessaires :**
Dans la première cellule, importez les bibliothèques `matplotlib.pyplot` pour les graphiques et `random` pour la génération de nombres aléatoires.


```python
import matplotlib.pyplot as plt
import random
```

## Étape 2 : Création de la liste de notes

**Définir une liste de notes :**
- Créez une liste contenant les notes des élèves. Affichez les.

```python
notes = [random.randint(0, 20) for i in range(40)]
print(notes)
```

## Étape 3 : Calcul des statistiques

**Écrire une fonction pour calculer les statistiques :**

- Créez une fonction `statistiques_notes` qui prend en entrée une liste de notes et retourne la moyenne, la note maximale et la note minimale.
- Utilisez une boucle `for` pour parcourir la liste et calculer ces valeurs.

```python
def statistiques_notes(notes):
    somme_notes = 0
    note_max = notes[0]
    note_min = notes[0]

    for note in notes:
        somme_notes += note
        if note > note_max:
            note_max = note
        if note < note_min:
            note_min = note

    moyenne = somme_notes / len(notes)
    return moyenne, note_max, note_min
```

**Écrire une fonction pour calculer les statistiques :**
- Appelez la fonction avec votre liste de notes et affichez les résultats.

```python
moyenne, note_max, note_min = statistiques_notes(notes)
print(f"La moyenne des notes est : {moyenne}")
print(f"La note la plus élevée est : {note_max}")
print(f"La note la plus basse est : {note_min}")
```

## Étape 4 : Visualisation des données

**Créer un histogramme des notes :**
- Utilisez `matplotlib` pour créer un histogramme des notes.
- Ajoutez des titres et des labels aux axes pour rendre le graphique plus lisible.

```python
plt.figure(figsize=(10, 5))
plt.hist(notes, bins=range(8, 21), edgecolor='black', alpha=0.7)
plt.title('Histogramme des notes')
plt.xlabel('Notes')
plt.ylabel('Fréquence')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

**Créer un graphique en barres des statistiques :**
- Créez un graphique en barres pour afficher la moyenne, la note maximale, et la note minimale.

```python
statistiques = ['Moyenne', 'Note Max', 'Note Min']
valeurs = [moyenne, note_max, note_min]

plt.figure(figsize=(8, 5))
plt.bar(statistiques, valeurs, color=['blue', 'green', 'red'])
plt.title('Statistiques des notes')
plt.ylabel('Valeurs')
plt.ylim(0, 20)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

## Étape 5 : Encapsulation dans une fonction

**Créer une fonction pour afficher les graphiques :**
- Écrivez une fonction `afficher_graphiques` qui prend en entrée une liste de notes et génère les graphiques.

```python
def afficher_graphiques(notes):
    moyenne, note_max, note_min = statistiques_notes(notes)

    # Histogramme des notes
    plt.figure(figsize=(10, 5))
    plt.hist(notes, bins=range(8, 21), edgecolor='black', alpha=0.7)
    plt.title('Histogramme des notes')
    plt.xlabel('Notes')
    plt.ylabel('Fréquence')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

    # Graphique en barres des statistiques
    statistiques = ['Moyenne', 'Note Max', 'Note Min']
    valeurs = [moyenne, note_max, note_min]

    plt.figure(figsize=(8, 5))
    plt.bar(statistiques, valeurs, color=['blue', 'green', 'red'])
    plt.title('Statistiques des notes')
    plt.ylabel('Valeurs')
    plt.ylim(0, 20)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()
```

**Tester la fonction :**
- Appelez la fonction `afficher_graphiques` avec votre liste de notes pour vérifier que tout fonctionne correctement.

```python
afficher_graphiques(notes)
```
