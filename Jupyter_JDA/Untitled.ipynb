```python
import matplotlib.pyplot as plt
import random
```

<div style="display: flex; align-items: flex-start;">
    <div>
        <h2>Étape 2 : Création de la liste de notes</h2>
        <p><strong>Définir une liste de notes :</strong></p>
        <ul>
            <li>Créez une liste contenant les notes des élèves. Affichez les.</li>
    </div>
</div>


```python
notes = [random.randint(0, 20) for i in range(40)]
print(notes)
```

<div style="display: flex; align-items: flex-start;">
    <div>
        <h2>Étape 3 : Calcul des statistiques</h2>
        <p><strong>Écrire une fonction pour calculer les statistiques :</strong></p>
        <ul>
            <li>Créez une fonction <code>statistiques_notes</code> qui prend en entrée une liste de notes et retourne la moyenne, la note maximale, et la note minimale.</li>
            <li>Utilisez une boucle <code>for</code> pour parcourir la liste et calculer ces valeurs.</li>
    </div>
</div>

```python
def statistiques_notes(notes):
    somme_notes = 0
    note_max = notes[0]
    note_min = notes[0]

    for note in notes:
        somme_notes += note
        if note > note_max:
            note_max = note
        if note < note_min:
            note_min = note

    moyenne = somme_notes / len(notes)
    return moyenne, note_max, note_min
```

<p><strong>Écrire une fonction pour calculer les statistiques :</strong></p>
 <li>Appelez la fonction avec votre liste de notes et affichez les résultats.</li>

```python
moyenne, note_max, note_min = statistiques_notes(notes)
print(f"La moyenne des notes est : {moyenne}")
print(f"La note la plus élevée est : {note_max}")
print(f"La note la plus basse est : {note_min}")
```

<div style="display: flex; align-items: flex-start;">
    <div>
        <h2>Étape 4 : Visualisation des données</h2>
        <p><strong>Créer un histogramme des notes :</strong></p>
        <ul>
            <li>Utilisez <code>matplotlib</code> pour créer un histogramme des notes.</li>
            <li>Ajoutez des titres et des labels aux axes pour rendre le graphique plus lisible.</li>
    </div>
</div>

```python
plt.figure(figsize=(10, 5))
plt.hist(notes, bins=range(8, 21), edgecolor='black', alpha=0.7)
plt.title('Histogramme des notes')
plt.xlabel('Notes')
plt.ylabel('Fréquence')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

<p><strong>Créer un graphique en barres des statistiques :</strong></p>
 <li>Créez un graphique en barres pour afficher la moyenne, la note maximale, et la note minimale.</li>

```python
statistiques = ['Moyenne', 'Note Max', 'Note Min']
valeurs = [moyenne, note_max, note_min]

plt.figure(figsize=(8, 5))
plt.bar(statistiques, valeurs, color=['blue', 'green', 'red'])
plt.title('Statistiques des notes')
plt.ylabel('Valeurs')
plt.ylim(0, 20)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

<div style="display: flex; align-items: flex-start;">
    <div>
        <h2>Étape 5 : Encapsulation dans une fonction</h2>
        <p><strong>Créer une fonction pour afficher les graphiques :</strong></p>
        <ul>
            <li>Écrivez une fonction <code>afficher_graphiques</code> qui prend en entrée une liste de notes et génère les graphiques.</li>
    </div>
</div>

```python
def afficher_graphiques(notes):
    moyenne, note_max, note_min = statistiques_notes(notes)

    # Histogramme des notes
    plt.figure(figsize=(10, 5))
    plt.hist(notes, bins=range(8, 21), edgecolor='black', alpha=0.7)
    plt.title('Histogramme des notes')
    plt.xlabel('Notes')
    plt.ylabel('Fréquence')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

    # Graphique en barres des statistiques
    statistiques = ['Moyenne', 'Note Max', 'Note Min']
    valeurs = [moyenne, note_max, note_min]

    plt.figure(figsize=(8, 5))
    plt.bar(statistiques, valeurs, color=['blue', 'green', 'red'])
    plt.title('Statistiques des notes')
    plt.ylabel('Valeurs')
    plt.ylim(0, 20)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()
```

<p><strong>Tester la fonction :</strong></p>
 <li>Appelez la fonction afficher_graphiques avec votre liste de notes pour vérifier que tout fonctionne correctement.</li>

```python
afficher_graphiques(notes)
```
